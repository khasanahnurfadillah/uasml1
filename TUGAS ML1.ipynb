```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_excel('dataset.xlsx')
```

```python
# Line Plot 1: Total luas panen padi per tahun
plt.figure()
df.groupby('tahun')['luas_panen_padi_total'].sum().plot()
plt.title('Total Luas Panen Padi per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Luas Panen Padi')
plt.show()
```

```python
# Line Plot 2: Rata-rata luas panen padi per tahun
plt.figure()
df.groupby('tahun')['luas_panen_padi_total'].mean().plot()
plt.title('Rata-rata Luas Panen Padi per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Luas Panen Padi')
plt.show()
```

```python
# Box Plot 1: Luas panen padi keseluruhan
plt.figure()
df.boxplot(column='luas_panen_padi_total')
plt.title('Box Plot Luas Panen Padi')
plt.show()
```

```python
# Box Plot 2: Luas panen padi per tahun
plt.figure()
df.boxplot(column='luas_panen_padi_total', by='tahun')
plt.title('Box Plot Luas Panen Padi per Tahun')
plt.suptitle('')
plt.xlabel('Tahun')
plt.ylabel('Luas Panen Padi')
plt.show()
```

```python
# Pie Chart 1: Distribusi data berdasarkan tahun
plt.figure()
df['tahun'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribusi Data Berdasarkan Tahun')
plt.ylabel('')
plt.show()
```

```python
# Pie Chart 2: Distribusi kategori luas panen padi
kategori = pd.cut(
    df['luas_panen_padi_total'],
    bins=3,
    labels=['Rendah', 'Sedang', 'Tinggi']
)

plt.figure()
kategori.value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribusi Kategori Luas Panen Padi')
plt.ylabel('')
plt.show()
```

```python
# 4. SCATTER PLOT
# =============================

plt.figure()
plt.scatter(df['tahun'], df['luas_panen_padi_total'])
plt.title('Scatter Plot Tahun vs Luas Panen Padi')
plt.xlabel('Tahun')
plt.ylabel('Luas Panen Padi')
plt.show()
```

```python
# 5. CORRELATION MATRIX
# =============================

corr = df[['luas_panen_padi_total', 'tahun']].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.show()
```

# Modeling

## Modeling classification / clustering

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_excel('dataset.xlsx')

# =============================
# 1. Seleksi Fitur
# =============================
X = df[['luas_panen_padi_total', 'tahun']]

# =============================
# 2. Normalisasi Data
# =============================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =============================
# 3. Penerapan K-Means Clustering
# =============================
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# =============================
# 4. Menampilkan Hasil
# =============================
df[['nama_kabupaten_kota', 'luas_panen_padi_total', 'tahun', 'Cluster']].head()
```
